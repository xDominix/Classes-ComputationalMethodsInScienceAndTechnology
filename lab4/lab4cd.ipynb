### 3. Korzystając ze wsparcia dla aproksymacji Pade w pakiecie Polynomials pokazać dowolny przykład (wraz z wykresem), gdzie przybliżenie jest lepsze od Taylora:

#### Poniżej funkcja, którą się zajmę:

```julia
function myfun(x) 
    return (log(1+x))/x
end
```

#### Wybrane punkty tej funkcji:

```julia
xs = 0:10
ys = map(myfun, xs)
```

#### Aproksymacja Taylora:

```julia
myfun_taylor = myfun(t)
myfun_t_pol = Polynomial(myfun_taylor.coeffs)
```

```julia
using Plots
default(fmt = :png)
plot(myfun_t_pol,  extrema(xs)..., label="aproksymacja Taylora")
scatter!(xs, ys, label ="dane punkty")
```

#### Aproksymacja Pade'go:

```julia
using Polynomials
p=Polynomials.PolyCompat.PadeApproximation.Pade(myfun_t_pol, 2,2)
```
